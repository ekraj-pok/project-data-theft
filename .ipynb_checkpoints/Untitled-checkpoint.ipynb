```python
import pandas as pd
```

```python
# import temperature data from kaggle : "https://www.kaggle.com/datasets/sudalairajkumar/daily-temperature-of-major-cities"
temperature_data = pd.read_csv("city_temperature.csv", low_memory = False)

# Remove countries before 2005 to limit the time series
temperature_data = temperature_data[~temperature_data['Year'].isin([201, 200,1995, 1996, 1997, 1998, 1999, 2000, 2001, 2002, 2003, 2004, 2005])]

# Convert farenheit in to celcious
temperature_data["average"] = temperature_data["AvgTemperature"].apply(lambda x: (x - 32) * 5 / 9)

# average temperature of countries over 15 years from 2006 to 2020
temperature_df = temperature_data.groupby("Country").agg({"AvgTemperature":"mean"}).reset_index()

# export to csv 
temperature_df.to_csv("temperature.csv")
```
