```python
import requests

url = "http://api.worldbank.org/v2/country?per_page=500&format=json"
response = requests.get(url)
countries = response.json()[1]
country_dict = {}
for country in countries:
    
    country_dict[country['id']] = country['name'], country['region']['value'],country["capitalCity"],country["incomeLevel"]["value"]
   
    
```

```python
df = pd.DataFrame(country_dict).T
```

```python
high = df[df[1] == "Europe & Central Asia"]
```
